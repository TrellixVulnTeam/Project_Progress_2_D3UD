# Building_classification_dataset

This file contain the code implemented to generate classification model from the annotation csv file of the indoor scene image dataset. Two types of datasets were generated:
 - Dataset 1: contains the information about the occurrence of the object detected.
 - Dataset 2: contains the information about the occurrence of the object detected and its respective geometrical position.
 
 Two methods were implemented two generate the about two dataset

## 1.2- Method 1: Dataset that contain the occurrence information and the geometrical position of the object detected

```python
kl=[]
for col in range(4*len(ls)):
    c= "val_"+ str(col)
    kl.append(c)
kl
```

```python
df_nw_pos = pd.DataFrame(columns=[kl])
```

```python
df_nw_pos['Class']=''
```

```python
df_nw_pos
```

### Main "For loop" for building the geometric position dataset

```python
i=0
for item in df_all.image:
    if i==0 or item!=item_prev:
        df_nw_pos.loc[i]=0
        dl=df_all[df_all['image']== item].index
        k=[df_all.loc[dls,"label"] for dls in dl]
        res = {k[s]: dl[s] for s in range(len(dl))}
        j=0
        for dls in dl:
            for j in k:
            
                f=ls.index(j)
                temp1=df_all.loc[res[j],"xmin"]
                df_nw_pos.iloc[i,4*f] = temp1


                temp1=df_all.loc[res[j],"ymin"]
                df_nw_pos.iloc[i,(4*f)+1] = temp1


                temp1=df_all.loc[res[j],"xmax"]
                df_nw_pos.iloc[i,(4*f)+2] = temp1


                temp1=df_all.loc[res[j],"ymax"]
                df_nw_pos.iloc[i,(4*f)+3] = temp1

            temp=df_all.loc[dl[0],"Class"] 
            df_nw_pos.loc[i,"Class"]=temp  
        item_prev=item
        i+=1      
        
df_nw_pos        

            
```
